```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Retail Sales Analysis_utf.csv')

# Display the first few rows
display(df.head())

# Display the information of the DataFrame
display(df.info())
```

```python
# Impute missing values with the mean
for col in ['age', 'quantiy', 'price_per_unit', 'cogs', 'total_sale']:
    df[col].fillna(df[col].mean(), inplace=True)

# Verify that missing values have been handled
display(df.info())
```

```python
from sklearn.preprocessing import StandardScaler

# Select relevant numerical features for clustering
features_for_clustering = ['age', 'quantiy', 'price_per_unit', 'cogs', 'total_sale']
X = df[features_for_clustering]

# Scale the selected numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store the preprocessed and scaled data in a DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=features_for_clustering)

# Display the first few rows of the scaled data
display(X_scaled_df.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Create empty lists to store inertia and silhouette scores
inertia = []
silhouette_scores = []

# Iterate through a range of possible numbers of clusters
k_range = range(2, 11)
for k in k_range:
    # Instantiate and fit the KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Set n_init to suppress warning
    kmeans.fit(X_scaled_df)

    # Append inertia
    inertia.append(kmeans.inertia_)

    # Calculate and append silhouette score
    score = silhouette_score(X_scaled_df, kmeans.labels_)
    silhouette_scores.append(score)

# Plot the elbow method graph
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)

# Plot the silhouette scores graph
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)

plt.tight_layout()
plt.show()
```

```python
# Define the features used for clustering
features_for_clustering = ['age', 'quantiy', 'price_per_unit', 'cogs', 'total_sale']

# Group the DataFrame by 'cluster_label' and calculate the mean of the clustering features
cluster_characteristics = df.groupby('cluster_label')[features_for_clustering].mean()

# Display the calculated mean values for each cluster
display(cluster_characteristics)

# Interpret the clusters based on the calculated means
print("Cluster Characteristics Interpretation:")
for cluster_id, characteristics in cluster_characteristics.iterrows():
    print(f"\nCluster {cluster_id}:")
    print(f"- Average Age: {characteristics['age']:.2f}")
    print(f"- Average Quantity: {characteristics['quantiy']:.2f}")
    print(f"- Average Price per Unit: {characteristics['price_per_unit']:.2f}")
    print(f"- Average COGS: {characteristics['cogs']:.2f}")
    print(f"- Average Total Sale: {characteristics['total_sale']:.2f}")
    # Provide a brief interpretation based on the values
    if characteristics['total_sale'] > df['total_sale'].mean() * 1.5:
        print("  - High-spending customers")
    elif characteristics['total_sale'] < df['total_sale'].mean() * 0.5:
         print("  - Low-spending customers")
    else:
        print("  - Average-spending customers")

    if characteristics['quantiy'] > df['quantiy'].mean() * 1.5:
        print("  - Purchase high quantities")
    elif characteristics['quantiy'] < df['quantiy'].mean() * 0.5:
        print("  - Purchase low quantities")
    else:
        print("  - Purchase average quantities")

    if characteristics['age'] > df['age'].mean() + df['age'].std():
        print("  - Older age group")
    elif characteristics['age'] < df['age'].mean() - df['age'].std():
        print("  - Younger age group")
    else:
        print("  - Middle age group")
```

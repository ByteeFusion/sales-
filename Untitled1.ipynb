```python
import pandas as pd
import io

df = pd.read_csv(io.StringIO(uploaded['Retail Sales Analysis_utf.csv'].decode('utf-8')))
display(df.head())
```

```python
print("Dataset Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
display(df.head())
```

```python
if 'sale_date' in df.columns:
    df['sale_date'] = pd.to_datetime(df['sale_date'], errors='coerce')
```

```python
print("\nSummary Statistics:\n", df.describe())
```

```python
if 'total_sale' in df.columns:
    total_sales = df['total_sale'].sum()
else:
    # Calculate total sales from quantity and price if 'total_sale' is not available
    if 'quantiy' in df.columns and 'price_per_unit' in df.columns:
      total_sales = (df['quantiy'] * df['price_per_unit']).sum()
    else:
      total_sales = 0 # Or handle this case appropriately

if 'total_sale' in df.columns:
    revenue_per_category = df.groupby('category')['total_sale'].sum()
else:
    # Calculate revenue per category from quantity and price if 'total_sale' is not available
    if 'quantiy' in df.columns and 'price_per_unit' in df.columns:
        revenue_per_category = df.groupby('category').apply(lambda x: (x['quantiy'] * x['price_per_unit']).sum())
    else:
        revenue_per_category = pd.Series(dtype=float) # Or handle this case appropriately


# Check if 'transactions_id' exists before calculating average order value
if 'transactions_id' in df.columns and total_sales > 0:
    average_order_value = total_sales / df['transactions_id'].nunique()
else:
    average_order_value = 0 # Or handle this case appropriately


print(f"\nTotal Sales: {total_sales:,.2f}")
print("\nRevenue per Category:\n", revenue_per_category)
print(f"\nAverage Order Value: {average_order_value:,.2f}")
```

```python
if 'sale_date' in df.columns:
    sales_over_time = df.groupby(df['sale_date'].dt.to_period('M')).apply(lambda x: (x['quantiy'] * x['price_per_unit']).sum()).reset_index(name='sales')
    sales_over_time['sale_date'] = sales_over_time['sale_date'].dt.to_timestamp()

    plt.figure(figsize=(10,5))
    sns.lineplot(data=sales_over_time, x='sale_date', y='sales', marker='o')
    plt.title("Sales Over Time")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.show()
```

```python
if 'sale_date' in df.columns:
    df['month'] = df['sale_date'].dt.month
    monthly_sales = df.groupby('month').apply(lambda x: (x['quantiy'] * x['price_per_unit']).sum()).reset_index(name='sales')

    plt.figure(figsize=(8,5))
    sns.barplot(data=monthly_sales, x='month', y='sales')
    plt.title("Average Sales by Month")
    plt.xlabel("Month")
    plt.ylabel("Sales")
    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=revenue_per_category.index, y=revenue_per_category.values)
plt.title("Revenue per Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```
